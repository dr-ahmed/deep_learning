```python
from PIL import Image
from deepface import DeepFace as df
from matplotlib import pyplot as plt
```

```python
img = plt.imread('img1.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
face = df.detectFace(img_path = 'img1.jpg')
plt.imshow(face)
plt.axis('off')
plt.show()
```

```python
try:
  face = df.detectFace(img_path = 'img2.jpg', enforce_detection=False)
  plt.imshow(face)
  plt.axis('off')
except Exception as e:
  print(type(e).__name__, ':', e)
```

```python
fig, axes = plt.subplots(1, 4, figsize = (20, 20))
img_list = ['img1.jpg', 'img6.jpg', 'img7.jpg', 'img8.jpg']
for i, ax in enumerate(axes.flat):
  ax.axis('off')
  ax.set_title(img_list[i])
  ax.imshow(Image.open(img_list[i]).resize((800,800)))
```
